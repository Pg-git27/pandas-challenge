## District Summary


```python
# Create a high level snapshot (in table form) of the district's key metrics, including:
# Total Schools
# Total Students
# Total Budget
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)
```

```python
df_schools = pd.read_csv("../Resources/schools_complete.csv")
df_schools.head()
```

```python
df_students = pd.read_csv("../Resources/students_complete.csv")
df_students.head()
```

```python
# Combine the data into a single dataset
join_students_data = pd.merge(
    df_schools, df_students, how="left", on=["school_name", "school_name"]
)
join_students_data
```

```python
# total number of schools
total_number_schools = len(join_students_data["School ID"].unique())
total_number_schools
```

```python
# total number of students
total_number_students = len(join_students_data["Student ID"].unique())
total_number_students
```

```python
# total budget
total_budget = df_schools["budget"].sum()
total_budget
```

```python
# average Math Score
average_math_score = df_students["math_score"].mean()
average_math_score
```

```python
# average reading score
average_reading_score = df_students["reading_score"].mean()
average_reading_score
```

```python
# percentage of students with a passing math score (70 or greater)
passing_math = df_students.loc[df_students["math_score"] >= 70, :][
    "student_name"
].count()
percentage_passing = (passing_math / total_number_students) * 100
percentage_passing
```

```python
# percentage of students with a passing reading score (70 or greater)
passing_reading = df_students.loc[df_students["reading_score"] >= 70, :][
    "student_name"
].count()
percentage_reading = (passing_reading / total_number_students) * 100
percentage_reading
```

```python
# overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (percentage_passing + percentage_reading) / 2
overall_passing_rate
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame(
    {
        "Total Schools": total_number_schools,
        "Total Students": total_number_students,
        "Total Budget": total_budget,
        "Average Math Score": average_math_score,
        "Average Reading Score": average_reading_score,
        "% Passing Math": percentage_passing,
        "% Passing Reading": percentage_reading,
        "% Overall Passing Rate": overall_passing_rate,
    },
    index=[0],
)

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map(
    "$ {:,.2f}".format
)

district_summary_df = district_summary_df[
    [
        "Total Schools",
        "Total Students",
        "Total Budget",
        "Average Math Score",
        "Average Reading Score",
        "% Passing Math",
        "% Passing Reading",
        "% Overall Passing Rate",
    ]
]

district_summary_df
```

## School Summary

```python
# Create an overview table that summarizes key metrics about each school, including:
# School Name
# School Type
# Total Students
# Total School Budget
# Per Student Budget
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)
```

```python
join_students_data["passing_math"] = join_students_data["math_score"] >= 70
join_students_data["passing_reading"] = join_students_data["reading_score"] >= 70

join_students_data
```

```python
# Use groupby by school_name

school_groupby_newtable = join_students_data.groupby(["school_name"]).mean()
school_groupby_newtable
```

```python
school_groupby_newtable["Per Student Budget"] = (
    school_groupby_newtable["budget"] / school_groupby_newtable["size"]
)
school_groupby_newtable["% Passing Math"] = round(
    school_groupby_newtable["passing_math"] * 100, 2
)
school_groupby_newtable["% Passing Reading"] = round(
    school_groupby_newtable["passing_reading"] * 100, 2
)
school_groupby_newtable["% Overall Passing Rate"] = round(
    (
        (
            school_groupby_newtable["passing_math"]
            + school_groupby_newtable["passing_reading"]
        )
        / 2
    )
    * 100,
    3,
)

# Merge with school_data to collect information about the type, size and budget
school_data_summary = pd.merge(
    school_groupby_newtable, df_schools, how="left", on=["school_name", "school_name"]
)
del school_data_summary["size_y"]
del school_data_summary["budget_y"]
del school_data_summary["Student ID"]
del school_data_summary["School ID_x"]

# Create a dataframe to store the results
school_summary_df = pd.DataFrame(
    {
        "School Name": school_data_summary["school_name"],
        "School Type": school_data_summary["type"],
        "Total Students": school_data_summary["size_x"],
        "Total School Budget": school_data_summary["budget_x"],
        "Per Student Budget": school_data_summary["Per Student Budget"],
        "Average Math Score": round(school_data_summary["math_score"], 2),
        "Average Reading Score": round(school_data_summary["reading_score"], 2),
        "% Passing Math": school_data_summary["% Passing Math"],
        "% Passing Reading": school_data_summary["% Passing Reading"],
        "% Overall Passing Rate": school_data_summary["% Overall Passing Rate"],
    }
)

# Formatting
school_summary_df["Total Students"] = school_summary_df["Total Students"].map(
    "{:,.0f}".format
)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map(
    "${:,.2f}".format
)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map(
    "${:,.2f}".format
)
school_summary_df
```

## Math scores by grade

```python
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
# Create a pandas series for each grade. Hint: use a conditional statement.
# Group each series by school
# Combine the series into a dataframe
# Optional: give the displayed data cleaner formatting
```

```python
# join_students_data
```

```python
# Create a pandas series for each grade. Group each series by school.

group_ninth_grade = (
    join_students_data[join_students_data["grade"] == "9th"]
    .groupby("school_name")
    .mean()["math_score"]
)
group_tenth_grade = (
    join_students_data[join_students_data["grade"] == "10th"]
    .groupby("school_name")
    .mean()["math_score"]
)
group_eleventh_grade = (
    join_students_data[join_students_data["grade"] == "11th"]
    .groupby("school_name")
    .mean()["math_score"]
)
group_twelveth_grade = (
    join_students_data[join_students_data["grade"] == "12th"]
    .groupby("school_name")
    .mean()["math_score"]
)

# Combine the series into a dataframe
df_mathscores_grade = pd.DataFrame(
    {
        "Ninth Grade": group_ninth_grade,
        "Tenth Grade": group_tenth_grade,
        "Eleventh Grade": group_eleventh_grade,
        "Twelveth Grade": group_twelveth_grade,
    }
)

# Optional formatting: Give the displayed data cleaner formatting
df_mathscores_grade[
    ["Ninth Grade", "Tenth Grade", "Eleventh Grade", "Twelveth Grade"]
] = df_mathscores_grade[
    ["Ninth Grade", "Tenth Grade", "Eleventh Grade", "Twelveth Grade"]
].applymap(
    "{:.2f}".format
)
df_mathscores_grade
```
